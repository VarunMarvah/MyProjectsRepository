# Flight Speed Analysis

```python
flights_input_speed=np.floor((flights_input['Distance']/flights_input['Air_Time'])*60)
flights_input_speed
```

```python
#Add new column in existing dataframe
flights_input['Aircraft_speed']=flights_input_speed
```

```python
flights_input.head(1)
```

```python
flights_input["Aircraft_speed"].describe()
```

```python
def CalculateFlightsSpeedRange(fromrange,torange):
    MaskRange=(flights_input["Aircraft_speed"]>=fromrange) & (flights_input["Aircraft_speed"]<=torange)
    dff=flights_input[MaskRange]
    return dff["Aircraft_speed"].count()
```

```python
df100=CalculateFlightsSpeedRange(0,100)
df200=CalculateFlightsSpeedRange(101,200)
df300=CalculateFlightsSpeedRange(201,300)
df400=CalculateFlightsSpeedRange(301,400)
df500=CalculateFlightsSpeedRange(401,500)
df600=CalculateFlightsSpeedRange(501,600)
df700=CalculateFlightsSpeedRange(601,700)
df800=CalculateFlightsSpeedRange(701,800)
df100
```

```python
flightsrange = [{'100': df100, '200': df200, '300':df300 , '400': df400 ,  '500': df500 , '600': df600 , '700': df700 , '800': df800}
        ]
mydf = pd.DataFrame(flightsrange)
```
